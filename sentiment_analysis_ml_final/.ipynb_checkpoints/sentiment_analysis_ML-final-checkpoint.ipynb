From the pie chart it can be seen that 55.9% of the total data has a rating of 5 while only 5.69% of the data has a rating of 2

<b> Let's bucketize the rating into 3 categories, namely: Positive, Neutral and Negative.<br>
    Rating 5 will be marked as Positive<br>
    Rating 3 & 4 will be bucktized together as Neutral<br>
    Rating 2 & 1 will be now be bucktized together as Negative

```python
def create_bucket(data):
    
    if data == 5:
        return "Positive"
    elif data == 4 or data == 3:
        return "Neutral"
    else:
        return "Negative"
```

```python
df['overall'] = df['overall'].apply(lambda x: create_bucket(x))
```

```python
df['overall'].value_counts()
```

<b>Let's generate the word cloud for the text

```python
#Wordcloud Representation

from wordcloud import STOPWORDS,WordCloud

def wcloud(text,title=None,figure_size=(24.0,16.0)):
    stopwords = set(STOPWORDS)
#     stopwords = stopwords.union({'one', 'br', 'Po', 'th', 'sayi', 'fo', 'Unknown'})
    
    wordcloud = WordCloud(stopwords=stopwords,random_state = 42,width=800, 
                    height=400,).generate(str(text))
    
    plt.figure(figsize=figure_size)
    plt.title(title,fontdict={'size': 40,})
    plt.imshow(wordcloud)
```

### A variety of words are present in all of the classes, let compare the ngrams for each of them

```python
from collections import defaultdict

df_pos = df[df["overall"]=='Positive']
df_neu = df[df["overall"]=='Neutral']
df_neg = df[df["overall"]=='Negative']

def generate_ngrams(text, n_gram=1):
    token = [token for token in text.lower().split(" ") if token != "" if token not in STOPWORDS]
    ngrams = zip(*[token[i:] for i in range(n_gram)])
    return [" ".join(ngram) for ngram in ngrams]

def generate_bar(df,color):
    plot = go.Bar(x=df['word_count'].values[::-1],y=df['word'].values[::-1],
                 orientation = 'h',showlegend=False,marker=dict(color=color))
    return plot
```

```python
#Bar for positive sentiments
freq_dict = defaultdict(int)
for rev in df_pos['reviewText']:
    for word in generate_ngrams(rev,2):
        freq_dict[word]+=1
sorted_freq = pd.DataFrame(sorted(freq_dict.items(),key= lambda x:x[1])[::-1])
sorted_freq.columns = ['word', 'word_count']
plot1 = generate_bar(sorted_freq.head(50),'green')
```

```python
#Bar for Neutral sentiments
freq_dict = defaultdict(int)
for rev in df_neu['reviewText']:
    for word in generate_ngrams(rev,2):
        freq_dict[word]+=1
sorted_freq = pd.DataFrame(sorted(freq_dict.items(),key= lambda x:x[1])[::-1])
sorted_freq.columns = ['word', 'word_count']
plot2 = generate_bar(sorted_freq.head(50),'blue')
```

```python
#Bar for Negative sentiments
freq_dict = defaultdict(int)
for rev in df_neg['reviewText']:
    for word in generate_ngrams(rev,2):
        freq_dict[word]+=1
sorted_freq = pd.DataFrame(sorted(freq_dict.items(),key= lambda x:x[1])[::-1])
sorted_freq.columns = ['word', 'word_count']
plot3 = generate_bar(sorted_freq.head(50),'red')
```

```python
# Creating two subplots
fig = tools.make_subplots(rows=1, cols=3, vertical_spacing=0.04,
                          subplot_titles=["Frequent words in Positive Reviews", 
                                          "Frequent words in Neutral Reviews",
                                          "Frequent words in Negative Reviews"])
fig.append_trace(plot1, 1, 1)
fig.append_trace(plot2, 1, 2)
fig.append_trace(plot3, 1, 3)
fig['layout'].update(height=500, width=1200, paper_bgcolor='rgb(233,233,233)', title="Bi-gram Plots")
py.iplot(fig, filename='word-plots')
```

It's time to create some meta features and check out how they are distributed in the dataset:

* Number of words in the text
* Number of unique words in the text
* Number of characters in the text
* Number of stopwords
* Number of punctuations
* Number of upper case words
* Number of title case words
* Average length of the words  

```python
df_rev = df.copy()
```

```python
## Number of words
df_rev['word_count'] =  df_rev['reviewText'].apply(lambda x: len(str(x).split()))
#Number of unique words
df_rev['unique'] =  df_rev['reviewText'].apply(lambda x: len(set(str(x).split())))
#Number of characters
df_rev['char_count'] =  df_rev['reviewText'].apply(lambda x: len(str(x)))
# Number of stopwords
df_rev['stopwords'] =  df_rev['reviewText'].apply(lambda x: len([word for word in str(x).lower().split() if word in STOPWORDS]))
#Number of Puncuations
df_rev['punct'] =  df_rev['reviewText'].apply(lambda x: len([char for char in str(x) if char in string.punctuation]))
#Number of UpperCase
df_rev['upper'] =  df_rev['reviewText'].apply(lambda x: len([word for word in str(x).split() if word.isupper()]))
#Number of Title Case
df_rev['title'] =  df_rev['reviewText'].apply(lambda x: len([word for word in str(x).split() if word.istitle()]))
#Average length of word
#Number of characters
df_rev['word_len'] =  df_rev['reviewText'].apply(lambda x: np.mean([len(word) for word in str(x).split()]))
```

```python
df_rev = df_rev.drop(labels=['asin','helpful','reviewTime','reviewerID','reviewerName','summary','unixReviewTime'],
                    axis=1)
```

### Since this sentiment analysis is a multiclass classification problem with class imbalance, plotting confusion matrix and analyzing it will be the most appropriate metric to judge the performance of a model

```python
def plot_confusion(y_true, y_pred, encoder=le):
    
    df_cm = pd.DataFrame(confusion_matrix(y_true, y_pred), index = encoder.classes_,
                      columns = encoder.classes_)
    plt.figure(figsize = (7,5))
    sns.heatmap(df_cm, annot=True)
```

### Let's move ahead with TF-IDF Vectorzination and use those as a features for our model

#### Preprocessing the text

```python
le = LabelEncoder()

df['overall'] = le.fit_transform(df['overall'])
```

```python
y = df['overall']
X = df.drop(labels=['asin', 'helpful', 'overall','reviewTime','reviewerID','reviewerName','unixReviewTime'], axis=1)
```

Converting the all the text to consist only of small letters

```python
X['reviewText'] = X['reviewText'].apply(lambda x: x.lower())
X['summary'] = X['summary'].apply(lambda x: x.lower())
```

Removing Punctuations

```python
punct = "/-'?!.,#$%\'()*+-/:;<=>@[\\]^_`{|}~" + '""“”’' + '∞θ÷α•à−β∅³π‘₹´°£€\×™√²—–&'

def fix_punc(sentence, punct):
    for p in punct:
        sentence = sentence.replace(p, '')
    return sentence
```

```python
X['reviewText'] = X['reviewText'].apply(lambda x: fix_punc(x, punct))
X['summary'] = X['summary'].apply(lambda x: fix_punc(x, punct))
```

Extracting Summary and review text as two different datasets

```python
X_sum = X.drop(labels='reviewText', axis=1)
X_rev = X.drop(labels='summary', axis=1)
```

```python
X_train_sum, X_test_sum, y_train_sum, y_test_sum = train_test_split(X_sum, y, test_size=0.1, 
                                                    stratify=y, random_state=17)
```

```python
X_train_rev, X_test_rev, y_train_rev, y_test_rev = train_test_split(X_rev, y, test_size=0.1, 
                                                    stratify=y, random_state=17)
```

```python
X_train_rev_tf = tfidf0.transform(X_train_rev['reviewText'])
X_test_rev_tf = tfidf0.transform(X_test_rev['reviewText'])
```

__Observations Made:__

* Model performa best when we generate tf-idf for __a word, bi-gram and tri-gram__, hence we can say bi-grams and tri-grams are important features when deciding the sentiment, same as intution suggestes.

__Scope of improvement:__

* Since class imbalance is present we can further use __SMOTE(Synthetic Minority Oversampling Technique)__ so as to sample data in a more appropriate way which will help us in creating more robust model.

* __Stratified K-Folds__ can be used to evalute and fine-tune model in way that will make it more robust.
